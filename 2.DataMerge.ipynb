```python
import pickle

import pandas as pd
import numpy as np

path = 'CountyVaccine/RaceVaccineBackup/STATE_VACCINE_2021-03-27.p'

with open(path, 'rb' ) as f:
    STATE_VACCINE = pickle.load(f)
    
CountyRaceVaccine = pd.concat([v for k,v in STATE_VACCINE.items()])
print(CountyRaceVaccine.shape)
CountyRaceVaccine
```

```python
CountyRaceVaccine = CountyRaceVaccine.replace(['Suppressed', '-'], np.nan)

cols = ['State', 'County', 'Geographic Area Name', 'Rate_White',  'Rate_Black', 'Rate_Parity',
        'Numb_White',  'Numb_Black' ]

CountyRaceVaccine = CountyRaceVaccine[cols]
CountyRaceVaccine.columns = ['State', 'County', 'Geographic Area Name', 'Rate_White',  'Rate_Black', 'Rate_Disparity',
        'Numb_White',  'Numb_Black' ]

CountyRaceVaccine
```

# Merge All the Data

```python
MergedCounty1 = pd.merge(MergedCounty,  INTERNET_SUBSCRIPTIONS, on = 'Geographic Area Name', how = 'left')
MergedCounty2 = pd.merge(MergedCounty1, EDUCATION,  on = 'Geographic Area Name', how = 'left')
MergedCounty3 = pd.merge(MergedCounty2, EMPLOYMENT, on = 'Geographic Area Name', how = 'left')
MergedCounty4 = pd.merge(MergedCounty3, Traffic, on='Geographic Area Name', how = 'left')

MergedCounty5 = pd.merge(MergedCounty4, BroadBoardSpeed, on='GEOID', how = 'left')
MergedCounty6 = pd.merge(MergedCounty5, VaccineCoverageIndex, on='GEOID', how = 'left')
MergedCounty7 = pd.merge(MergedCounty6, VaccineFaclility, on='GEOID', how = 'left')
MergedCounty8 = pd.merge(MergedCounty7, COVID, on = 'GEOID', how = 'left')
MergedCounty9 = pd.merge(MergedCounty8, PoliticalIdeology, on='GEOID', how = 'left')

MergedCounty10 = pd.merge(MergedCounty9,  MeanIncome, on='GEOID', how = 'left')
MergedCounty11 = pd.merge(MergedCounty10, MedianIncome, on='GEOID', how = 'left')
MergedCounty12 = pd.merge(MergedCounty11, FCC, on='GEOID', how = 'left')
MergedCounty13 = pd.merge(MergedCounty12, HealthRanking, on='GEOID', how = 'left')
MergedCounty14 = pd.merge(MergedCounty13, ADI, on='GEOID', how = 'left')
MergedCounty15 = pd.merge(MergedCounty14, SDI, on='GEOID', how = 'left')
MergedCounty16 = pd.merge(MergedCounty15, AgeDisparity, on='GEOID', how = 'left')
MergedCounty17 = pd.merge(MergedCounty16, FluVax, on='GEOID', how = 'left')
MergedCounty18 = pd.merge(MergedCounty17, RaceBias, on='GEOID', how = 'left')
MergedCounty19 = pd.merge(MergedCounty18, SexBias, on='GEOID', how = 'left')
MergedCounty20 = pd.merge(MergedCounty19, SocialVulnerabilityIndex, on='GEOID', how = 'left')
MergedCounty21 = pd.merge(MergedCounty20, VaxHesitancy, on='GEOID', how = 'left')

print(MergedCounty21.shape)
print(list(MergedCounty21.columns))

# SocialVulnerabilityIndex
```
